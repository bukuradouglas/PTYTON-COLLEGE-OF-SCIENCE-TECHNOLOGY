```python
import numpy as np
import matplotlib.pyplot as plt

# 1. f(x) = x^3 - 3x, x in [-3, 3]
x1 = np.linspace(-3, 3, 400)
f1 = x1**3 - 3*x1

# 2. f(x) = ln(x^2 + e^(2x)), x in [-3, 3]
x2 = np.linspace(-3, 3, 400)
f2 = np.log(x2**2 + np.exp(2*x2))

# 3. f(x) = (x^3 + x^2 - x + 1)/(x - 2), x in [0, 4]
x3 = np.linspace(0, 4, 400)
f3 = (x3**3 + x3**2 - x3 + 1) / (x3 - 2)

# 4. f(θ) = cos(πθ), θ in [-2, 2]
theta = np.linspace(-2, 2, 400)
f4 = np.cos(np.pi * theta)

# Plotting all on the same graph
plt.figure(figsize=(16, 10))

plt.plot(x1, f1, label='f(x) = x³ - 3x')
plt.plot(x2, f2, label='f(x) = ln(x² + e^{2x})')
plt.plot(x3, f3, label='f(x) = (x³ + x² - x + 1)/(x - 2)')
plt.plot(theta, f4, label='f(θ) = cos(πθ)')

# Add labels and title
plt.xlabel('X / θ')
plt.ylabel('f(x) / f(θ)')
plt.title('Plots of All the Given Functions')
plt.axvline(x=2, color='gray', linestyle='--')  # show possible asymptote
plt.ylim(-40, 50)  # Adjust Y-axis limits to view everything well

plt.grid(True)
plt.legend()
plt.savefig("All in one function_plots.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. f(x) = x^3 - 3x, x ∈ [-3, 3]
x1 = np.linspace(-3, 3, 400)
f1 = x1**3 - 3*x1

# 2. f(x) = ln(x^2 + e^(2x)), x ∈ [-3, 3]
x2 = np.linspace(-3, 3, 400)
f2 = np.log(x2**2 + np.exp(2*x2))

# 3. f(x) = (x^3 + x^2 - x + 1)/(x - 2), x ∈ [0, 4]
x3 = np.linspace(0, 4, 400)
f3 = (x3**3 + x3**2 - x3 + 1) / (x3 - 2)

# 4. f(θ) = cos(πθ), θ ∈ [-2, 2]
theta = np.linspace(-2, 2, 400)
f4 = np.cos(np.pi * theta)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Plots of Each Functions", fontsize=16)

# Plot 1
axs[0, 0].plot(x1, f1, 'b')
axs[0, 0].set_title('f(x) = x³ - 3x')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].grid(True)

# Plot 2
axs[0, 1].plot(x2, f2, 'g')
axs[0, 1].set_title('f(x) = ln(x² + e^{2x})')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')
axs[0, 1].grid(True)

# Plot 3
axs[1, 0].plot(x3, f3, 'r')
axs[1, 0].set_title('f(x) = (x³ + x² - x + 1)/(x - 2)')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')
axs[1, 0].grid(True)
axs[1, 0].axvline(x=2, color='gray', linestyle='--')  # Mark asymptote

# Plot 4
axs[1, 1].plot(theta, f4, 'm')
axs[1, 1].set_title('f(θ) = cos(πθ)')
axs[1, 1].set_xlabel('θ')
axs[1, 1].set_ylabel('f(θ)')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("Separate function_plots.png", dpi=300)
plt.show()
```
